## plotEPCurvesExample.ipynb
This script plots EP curves generated in Remetrica financial modelling software, showing the effect of increasing correlation on aggregate (AEP) and occurrence (OEP) exceedance probability (or Return period) curves.

```python
from pathlib import Path
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the root path
rootPath = Path('C:/Users/alexa/Documents/02_work/02_start/02_deliv/05_pk_correlation/remetrica/plot')
```

## Load the OEP and AEP curves

```python
OEP = pd.read_csv(rootPath/'plotExampleOEP_DummyCustomSev.csv')
AEP = pd.read_csv(rootPath/'plotExampleAEP_DummyCustomSev.csv')
OEP=OEP.set_index('RP')
AEP=AEP.set_index('RP')
```

```python
OEP
```

## Plot the EP curves

```python
colors=[
(68/255,84/255,106/255),    
(254/255,106/255,70/255), 
(133/255,239/255,205/255),
(255/255,192/255,0/255),    
(165/255,165/255,165/255), 
(255/255,193/255,179/255), 
(184/255,193/255,218/255),    
(24/255,186/255,134/255)
]
```

```python
plotTitles=['Occurrence Exceedance Probability','Aggregate Exceedance Probability']

fig, axes = plt.subplots(1, 2,figsize=(10,4), constrained_layout=True)

# OEP
ax=axes[0]
ax.set_title(plotTitles[0], fontsize=16)
for i in range(len(OEP.columns)):
    ax.plot(OEP.index, OEP.iloc[:,i], marker='none', linestyle='--', color=colors[i], label=OEP.columns[i])
ax.set_ylabel('Exceedance Probability Loss', fontsize=14)
ax.set_xlabel('Return Period (years)', fontsize=14)
ax.set_xlim(0, 500)
ax.set_ylim(0, 2e8)
ax.tick_params(axis='both',labelsize=14)
ax.legend(loc='upper left')

# AEP
ax=axes[1]
ax.set_title(plotTitles[1], fontsize=16)
for i in range(len(AEP.columns)):
    ax.plot(AEP.index, AEP.iloc[:,i], marker='none', linestyle='--', color=colors[i], label=AEP.columns[i])
ax.set_ylabel('Exceedance Probability Loss', fontsize=14)
ax.set_xlabel('Return Period (years)', fontsize=14)
ax.set_xlim(0, 500)
ax.set_ylim(0, 2e8)
ax.tick_params(axis='both',labelsize=14)
ax.legend(loc='upper left')


plt.show()
```
