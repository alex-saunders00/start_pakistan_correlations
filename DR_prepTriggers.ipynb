## DR_prepTriggers.ipynb
This script prepares the data for Pakistan drought, to compute the triggers. It takes the raw data shared by Emily Black (University of Reading), computes the district populations and total Province populations to get the population thresholds. And it computes the 20%ile of the drought variable in each district using the historical data for 2001-2021 (20 years) after linear detrending. 

```python
from pathlib import Path
import os
import sys
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import date, timedelta
import matplotlib.pyplot as plt
from scipy import signal
from scipy import stats
```

```python
# Set the root path
rootPath = Path('C:/Users/alexa/Documents/02_work/02_start/02_deliv/05_pk_correlation/dr/data/')
```

## Load the population data

```python
popData=pd.read_csv(rootPath/'emily/demographic_updated.csv')
```

```python
# Aggregate from District to Province
popProvinces = pd.DataFrame(popData.groupby(by='Province').Rural_population.sum())
```

```python
# Threshold = 25% of province rural population
threshold=0.25
popProvinces['Province_threshold']=popProvinces.Rural_population*threshold
```

```python
outPath=rootPath/'prep'
outPath.mkdir(exist_ok=True)
```

```python
popProvinces.to_csv(outPath/'ProvinceThreshold.csv', index=True)
```

```python
distsPunjab = popData.District_id[popData.Province=='Punjab']
distsSindh = popData.District_id[popData.Province=='Sindh']
distsBaloch = popData.District_id[popData.Province=='Balochistan']
```

## Load the historical variables "observed" data

```python
provinces = list(popProvinces.index.values)
variables=['precip','yield','vhi']
inputFiles=['precip_hc_poi_21_30_init_31.csv',
           'all_yield_model_wheat_date_31_interannual.hea.csv',
           'vhi_hc_poi_21_30_init_31.csv']
```

```python
precipData=pd.read_csv(rootPath/('emily/'+inputFiles[0]))
yieldData=pd.read_csv(rootPath/('emily/'+inputFiles[1]))
vhiData=pd.read_csv(rootPath/('emily/'+inputFiles[2]))
```

### Balochistan - precip

```python
# Pivot the data
precipData=precipData[precipData.District_id.isin(distsBaloch)].pivot(index='District_id', columns='Year', values='Observed_precip')
```

```python
# Add the 20th percentile, for only last 20 years
pctile=0.2
yearsStart=2002
yearsEnd=2021
nYears=yearsEnd-yearsStart+1
print(nYears)
precipDataSub = precipData[precipData.columns[np.logical_and(precipData.columns>=yearsStart, precipData.columns<=yearsEnd)]]
precipDataSub['pctile']=precipDataSub.quantile(pctile, axis=1)
```

```python
precipDataSub.to_csv(outPath/'precipThreshold.csv', index=True)
```

### Punjab - yield

```python
# Pivot the data
yieldData=yieldData[yieldData.District_id.isin(distsPunjab)].pivot(index='District_id', columns='Year', values='Obs_yield')
```

```python
# Add the 20th percentile, for only last 20 years
pctile=0.2
yearsStart=1983
yearsEnd=2018
nYears=yearsEnd-yearsStart+1
print(nYears)
yieldDataSub = yieldData[yieldData.columns[np.logical_and(yieldData.columns>=yearsStart,yieldData.columns<=yearsEnd)]]
yieldDataSub['pctile']=yieldDataSub.quantile(pctile, axis=1)
```

```python
yieldDataSub
```

```python
yieldDataSub.to_csv(outPath/'yieldThreshold.csv', index=True)
```

#### Illustration of linear detrending

```python
fig, ax = plt.subplots()
x=np.array(list(yieldDataSub.columns[0:-1]))
y=np.array(list(yieldDataSub[yieldDataSub.index==601].values[0][0:-1]))
mask = ~np.isnan(x) & ~np.isnan(y)
y_detrend=signal.detrend(y[mask])
res = stats.linregress(x[mask], y[mask])
ypred = [res.intercept + res.slope*xi for xi in x]
ax.plot(x,y, label='data')
ax.plot(x,ypred, label='linear')
ax.plot(x[mask],y_detrend, label='detrended')
ax.legend()
plt.show()
```

### Use the detrended data as the historical dataset and to compute the percentile

```python
x=np.array(list(yieldDataSub.columns[0:-1]))
detrendedDF=pd.DataFrame(data=[], index=yieldDataSub.columns[0:-1])

for i, distID in enumerate(yieldDataSub.index):
    y=np.array(list(yieldDataSub[yieldDataSub.index==distID].values[0][0:-1]))
    mask = ~np.isnan(x) & ~np.isnan(y)
    y_detrend=signal.detrend(y[mask])
    xy=pd.DataFrame(data=[x[mask], y_detrend])
    xy.columns = xy.iloc[0]
    xy=xy.drop([0])
    
    detrendedDF=detrendedDF.merge(xy.T, left_index=True, right_index=True, how='outer')

detrendedDF.columns=yieldDataSub.index
detrendedDF=detrendedDF.T
detrendedDF['pctile']=detrendedDF.quantile(pctile,axis=1)
```

```python
detrendedDF
```

```python
detrendedDF.to_csv(outPath/'yieldThresholdDetrended.csv', index=True)
```
